```python
def VanderPol(x,t,p):
    return[x[1], p*(1-x[0]**2)*x[1]-x[0]]
```

```python
t = linspace(0,100,10001)
```

```python
xout = odeint(VanderPol,[1.0,0.0],t,args=(0.1,))
```

```python
plot(t,xout[:,0])
```

```python
plot(xout[:,0],xout[:,1])
```

```python
np.fft.rfft(xout[:,0])
```

```python
def fft_w(x,ffx):
    n = len(ffx)
    delta_x = x[1] - x[0]
    omega = linspace(0,1.0/delta_x, n/2 +1)
    omega = omega*pi
    T = np.fft.rfft(ffx)/n
    espectro = abs(T)
    return omega , espectro
```

```python
for i in linspace(0.0,10.0,10):
    xout = odeint(VanderPol,[1.0,0.0],t,args=(i,))
    plot(xout[:,0],xout[:,1])
```
